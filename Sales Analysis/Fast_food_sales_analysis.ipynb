```python
df= pd.read_csv("Balaji Fast Food Sales.csv")
df['date']= df['date'].astype(str).str.strip()

df['date_parsed'] = np.where(
    df['date'].str.contains('/'),
    pd.to_datetime(df['date'],format= '%m/%d/%Y', errors='coerce'),
    pd.to_datetime(df['date'],format= '%d-%m-%Y',errors= 'coerce')
)

df['date'] = df['date_parsed']
df.drop(columns=['date_parsed'],inplace= True)
```

```python
df.describe()
```

```python
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day_name()
df['year'] = df['date'].dt.year
df
```

```python

df.fillna({'transaction_type' : 'Unknown'}, inplace= True)
```

**Total Sales (KPI)**

```python
total_sales = df['transaction_amount'].sum()
print("Total Sales:", total_sales)
```

**Top Selling Items**


```python
item_sales= df.groupby('item_name')['transaction_amount'].sum().sort_values(ascending= False)
print(item_sales)
```

**Most Sold Item (Quantity)**

```python
df.groupby('item_name')['quantity'].sum().sort_values(ascending= False)
```

**Sales by Time of Day**

```python
df.groupby('time_of_sale')['transaction_amount'].sum().sort_values(ascending= False)
```
